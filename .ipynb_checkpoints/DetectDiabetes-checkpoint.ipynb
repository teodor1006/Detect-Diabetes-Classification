## Model Building

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns="Outcome")
y = df["Outcome"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 0)
print('X_train size: {}, X_test size: {}'.format(X_train.shape, X_test.shape))
```

# Standard Scaling

```python
import warnings
from sklearn.preprocessing import StandardScaler

warnings.filterwarnings("ignore", category=UserWarning)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Using GridSearchCV to find the best classification algorithm for this problem
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
def find_best_model(X, y):
    models = {
        'logistic_regression': {
            'model': LogisticRegression(class_weight="balanced"),
            'parameters': {
                'C': [0.1, 1, 5, 10, 100],
                'max_iter': [10000], 
            }
        },
        
        'decision_tree': {
            'model': DecisionTreeClassifier(splitter='best'),
            'parameters': {
                'criterion': ['gini', 'entropy'],
                'max_depth': [10, 15, 20, 25, 30],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4]
            }
        },
        
        'random_forest': {
            'model': RandomForestClassifier(criterion='entropy'),
            'parameters': {
                'n_estimators': [10, 15, 20, 50, 100, 200],
                'max_depth': [None, 10, 20, 30],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4]
            }
        },
        
        'svm': {
            'model': SVC(),
            'parameters': {
                'kernel': ["rbf", "linear"],
                "C": [0.1, 1, 10, 100],
                'gamma': ['scale', 0.001, 0.01, 0.1, 1]
            }
        }
    }
    
    scores = [] 
    cv_shuffle = ShuffleSplit(n_splits=5, train_size=0.8, random_state=0)
        
    for model_name, model_params in models.items():
        gs = GridSearchCV(model_params['model'], model_params['parameters'], cv=cv_shuffle, return_train_score=False, n_jobs=-1)
        gs.fit(X, y)
        scores.append({
            'model': model_name,
            'best_parameters': gs.best_params_,
            'score': gs.best_score_
        })
        
    return pd.DataFrame(scores, columns=['model','best_parameters','score'])

find_best_model(X_train, y_train)
```

Note: Well the SVM has the highest score, so we need to optimize the hyperparameters

```python
# Using cross_val_score for gaining average accuracy
from sklearn.model_selection import cross_val_score
scores = cross_val_score(SVC(), X_train, y_train, cv=5)
average_accuracy = scores.mean()

print('Average Accuracy: {:.2f}%'.format(average_accuracy * 100))
```

```python
classifier = SVC(C=1, kernel='linear', random_state=42, gamma='scale')
classifier.fit(X_train, y_train)
```

## Evaluating the model

```python
# Accuracy Score on test data
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test)
score = round(accuracy_score(y_test, y_pred),4)*100
print("Accuracy on test set: {:.2f}%".format(score))
```

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Accuracy Score on training data
y_train_pred = classifier.predict(X_train)
score = round(accuracy_score(y_train, y_train_pred),4)*100
print("Accuracy on training set: {:.2f}%".format(score))
```

```python
# Classification Report
print(classification_report(y_train, y_train_pred))
```

```python
# Plot the roc_curve
from sklearn.metrics import roc_curve, auc

y_scores = classifier.decision_function(X_test)
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(10, 7))
plt.plot(fpr, tpr, label='ROC curve (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

## Some Predictions

```python
# Creating a function for prediction
def predict_diabetes(Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DPF, Age):
    x = np.array([[int(Pregnancies), 
                   float(Glucose), 
                   float(BloodPressure), 
                   float(SkinThickness), 
                   float(Insulin), 
                   float(BMI), 
                   float(DPF), 
                   int(Age)]])
    x = sc.transform(x)
    prediction_model = classifier.predict(x)
    return prediction_model
```

```python
# Prediction 1
# Input sequence: Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DPF, Age
prediction = predict_diabetes(2, 81, 72, 15, 76, 30.1, 0.547, 25)[0]
if prediction:
  print('Oops! You have diabetes.')
else:
  print("Great! You don't have diabetes.")
```

# GUI

```python
from tkinter import *
import joblib

def show_entry_fields():
    input_data = [float(entry_vars[i].get()) for i in range(8)]
    model = joblib.load("Detect Diabetes")
    result = model.predict([input_data])
    
    result_label.config(text="Diabetic" if result == 1 else "Non-diabetic")

master = Tk()
master.title("Detecting Diabetes")

Label(master, text="Detecting Diabetes", bg="black", fg="white").grid(row=0, columnspan=2)

labels = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DPF", "Age"]

entry_vars = []

for i, label_text in enumerate(labels):
    Label(master, text=label_text).grid(row=i + 1)
    entry_var = Entry(master)
    entry_var.grid(row=i + 1, column=1)
    entry_vars.append(entry_var)

result_label = Label(master, text="")
result_label.grid(row=len(labels) + 1)

Button(master, text='Predict', command=show_entry_fields).grid()

mainloop()
```
